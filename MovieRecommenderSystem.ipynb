```python
import ast
```

```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_1(obj):
    L =[]
    count=0
    for i in ast.literal_eval(obj):
        
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L    
```

```python
movies['cast']=movies['cast'].apply(convert_1)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def director_fetch(obj):
    L =[]
    
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L    
```

```python
movies['crew']=movies['crew'].apply(director_fetch)
```

```python
movies.sample(2)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'][0]
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.sample(15)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

# Problems
- how to check which tags are similar (solution { make a vector of most frequent words in tags then find its cosine angle which methametically give us most similar tags })
- other one is none usable words like {is,are ,a the etc} solution({remove them from tag string})
- last one important words which have similar meaning have to replace by one word only ex(love,loved,loving) to (love)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
movie_index = new_df[new_df['title']=='Avatar'].index[0]
```

```python
movie_index
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommender(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommender("The Avengers")
```
